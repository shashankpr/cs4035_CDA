```python
df_ip = pd.read_table("C:\Users\Apourva\Downloads\ips.txt", header = None)
```

```python
ip_list = df_ip[0].tolist()
ip_frequency_count = Counter(ip_list)
```

```python
len(ip_list)
```

```python
# Top 10 most frequently occuring IP addresses
ip_frequency_count.most_common(10)
```

```python
def Frequent (k):
    c = {}
    T = []
    for ip in ip_list:
        if ip in T:
            c[ip] = c[ip] + 1
        else:
            if len(T) < k:
                T.append(ip)
                c[ip] = 1
            else:
                for j in T:
                    c[j] = c[j] - 1
                    if c[j] == 0:
                        T.remove(j)
    return T, c
```

```python
T_10, c_10 = Frequent ( 10 )
for i in T_10:
    print i, " : " , c_10[i]
```

According to the FREQUENT algorithm, all IP addresses which have frequency greater than n/k must be reported.
Here, n = 3414011 and k = 10 and n/k = 341401

The most commonly occuring IP address (115.177.11.215) only has frequency of 294690 which is less that 341401. Hence, we expect that the FREQUENT algorithm does not return any IP addresses.

However, we see that the algorithm returns 6 values.


```python
T_100, c_100 = Frequent ( 100 )
print "length of T_100 : ", len(T_100)
for i in T_100:
    print i, " : ", c_100[i]
```

```python
dict_100 = {}
for i in ip_frequency_count:
    if ip_frequency_count[i] > 34140:
        dict_100[i] = ip_frequency_count[i]
print dict_100
```

```python
T_1000, c_1000 = Frequent ( 1000 )
print "length of T_1000 : ", len(T_1000)
for i in T_1000:
    print i, " : ", c_1000[i]
```

```python
dict_1000 = {}
not_identified_by_algorithm = []
for i in ip_frequency_count:
    if ip_frequency_count[i] > 3414:
        dict_1000[i] = ip_frequency_count[i]
        if i not in T_1000:
            not_identified_by_algorithm.append(i)
print dict_1000
print "length : " , len(dict_1000)
print "IP addresses not reported by FREQUENT alforithm : ", not_identified_by_algorithm
```
