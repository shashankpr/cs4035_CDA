### Daily Trend

```python
df1 = normal_csv.groupby(pd.TimeGrouper('D'))
```

```python
df1['LIT101'].plot(figsize=(15,5))
```

### Hourly Trend

```python
one_day_data = df1.get_group('2015-12-22')
```

```python
one_day_data[0:10]
```
